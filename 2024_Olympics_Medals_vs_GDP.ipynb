## 6. Inferência Estatística

```python
# INTERVALO DE CONFIANÇA
import scipy.stats as stats

# Intervalo de confiança de 95% para o PIB médio
mean_gdp = data['gdp'].mean()
sem_gdp = stats.sem(data['gdp'])  # erro padrão da média
intervalo_gdp = stats.t.interval(0.95, len(data['gdp'])-1, loc=mean_gdp, scale=sem_gdp)
print("Intervalo de Confiança para PIB médio:", intervalo_gdp)
```

```python
# TESTE DE HIPÓTESE
# Dividir os dados em duas amostras
grupo1 = data[data['total'] > 10]['gdp']
grupo2 = data[data['total'] <= 10]['gdp']

# Teste t de hipóteses para duas amostras
t_stat, p_value = stats.ttest_ind(grupo1, grupo2)
print("Teste t - Estatística t:", t_stat, "P-valor:", p_value)
```

```python
# TESTE PARA TRÊS OU MAIS AMOSTRAS (ANOVA)
# ANOVA para comparar PIB entre regiões
anova_result = stats.f_oneway(data[data['region'] == 'Europa']['gdp'],
                              data[data['region'] == 'Ásia']['gdp'],
                              data[data['region'] == 'América']['gdp'])
print("ANOVA - Estatística F:", anova_result.statistic, "P-valor:", anova_result.pvalue)
```

```python
# TESTE DE CORRELAÇÃO E REFRESSÃO LINEAR SIMPLES
# Correlação de Pearson entre PIB e total de medalhas
correlacao, _ = stats.pearsonr(data['gdp'], data['total'])
print("Correlação entre PIB e Total de Medalhas:", correlacao)

# Regressão linear
from sklearn.linear_model import LinearRegression

X = data[['gdp']]
y = data['total']
modelo = LinearRegression()
modelo.fit(X, y)

print("Coeficiente da regressão:", modelo.coef_)
print("Intercepto:", modelo.intercept_)
```

```python
# TESTE QUI-QUADRADO
# Criar uma tabela de contingência
data['conquistou_mais_10'] = data['total'] > 10
contingencia = pd.crosstab(data['region'], data['conquistou_mais_10'])

# Teste Qui-Quadrado
chi2, p, dof, expected = stats.chi2_contingency(contingencia)
print("Qui-Quadrado - Estatística chi2:", chi2, "P-valor:", p)
```

## 7. Aprendizado de Máquina

### A. Aprendizado Supervisionado

```python
# VARIÁVEIS DEPENDENTES
data['medalha_obtida'] = data['total'].apply(lambda x: 1 if x > 0 else 0)

variaveis_independentes = ['gold', 'silver', 'bronze', 'gdp', 'population']
X = data[variaveis_independentes]
Y = data['medalha_obtida']

print(data[['country', 'medalha_obtida']].head())
```

```python
# DADOS DE TREINAMENTO E VALIDAÇÃO
from sklearn.model_selection import train_test_split
X_train, X_validation, Y_train, Y_validation = train_test_split(X, Y, test_size=0.15, random_state=30)
print(f'Treinamento: {X_train.shape}, Validação: {X_validation.shape}')
```

```python
# LISTA DE MODELOS DE CLASSIFICAÇÃO PARA AVALIAÇÃO
from sklearn.linear_model import LogisticRegression
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC

# Criar uma lista de modelos
models = []
models.append(('Regressão Logística', LogisticRegression(solver='liblinear')))
models.append(('Análise Discriminante Linear', LinearDiscriminantAnalysis()))
models.append(('K-Nearest Neighbors', KNeighborsClassifier()))
models.append(('Árvore de Decisão', DecisionTreeClassifier()))
models.append(('Naive Bayes', GaussianNB()))
models.append(('Máquina de Vetores de Suporte', SVC(gamma='auto')))

```

```python
# DESEMPENHO DOS MODELOS
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC

# Separar os dados em treino e validação com estratificação
X_train, X_validation, Y_train, Y_validation = train_test_split(
    X, Y, test_size=0.15, random_state=30, stratify=Y
)

# Verificar a distribuição das classes após a estratificação
print("Distribuição no treino:", Y_train.value_counts())
print("Distribuição na validação:", Y_validation.value_counts())

# Criar uma lista de modelos de classificação
models = [
    ('Regressão Logística', LogisticRegression(solver='liblinear')),
    ('Análise Discriminante Linear', LinearDiscriminantAnalysis()),
    ('K-Nearest Neighbors', KNeighborsClassifier()),
    ('Árvore de Decisão', DecisionTreeClassifier()),
    ('Naive Bayes', GaussianNB()),
    ('Máquina de Vetores de Suporte', SVC(gamma='auto'))
]

# Avaliar cada modelo
for name, model in models:
    model.fit(X_train, Y_train)  # Treinar o modelo
    predictions = model.predict(X_validation)  # Fazer previsões no conjunto de validação
    accuracy = accuracy_score(Y_validation, predictions)  # Calcular a acurácia
    print(f'{name}: Acurácia = {accuracy:.2f}')
```

```python
import pandas as pd
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.svm import SVR
import matplotlib.pyplot as plt
import numpy as np

# Convertendo variáveis categóricas em variáveis dummy (one-hot encoding)
data_encoded = pd.get_dummies(data, drop_first=True)

# Ajustar variáveis independentes (X) e dependentes (Y)
X = data_encoded.drop(columns=['total'])
Y = data_encoded['total']
```

```python
# Lista de modelos de regressão a serem avaliados
models = [
    ('Linear Regression', LinearRegression()),
    ('Decision Tree', DecisionTreeRegressor()),
    ('Random Forest', RandomForestRegressor()),
    ('K-Nearest Neighbors', KNeighborsRegressor()),
    ('Support Vector Regression', SVR())
]

# Avaliação dos modelos usando cross_val_score
results = []
acuracia = []
erro = []
nome = []
n = 0

for name, model in models:
    try:
        Models_results = cross_val_score(model, X, Y, cv=5, scoring='r2')
        results.append(Models_results)
        msg = "%s: %f (%f)" % (name, round(Models_results.mean(), 2), Models_results.std())
        acuracia.append(round(Models_results.mean(), 2))
        erro.append(Models_results.std())
        nome.append(name)
        print(msg)
    except ValueError as e:
        print(f"Erro ao treinar o modelo {name}: {e}")
```

```python
plt.figure(figsize=(10, 4))
plt.title('Figura 1: Acurácia (R²) dos Modelos de Regressão')

barWidth = 0.4
r = np.arange(len(nome))
a = plt.barh(nome, acuracia, color=['red', 'blue', 'green', 'purple', 'orange'], xerr=erro, edgecolor='black')

plt.bar_label(a, padding=3)
plt.xlabel('R² Score')
plt.ylabel('Modelos')
plt.grid(axis='x', linestyle='--')
plt.show()
```

```python
# Avaliação do Modelo de Regressão Logística
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score, roc_curve, auc
from sklearn.preprocessing import label_binarize
from sklearn.multiclass import OneVsRestClassifier
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

# Treinando o modelo de Regressão Logística com OneVsRestClassifier
logistic_model = OneVsRestClassifier(LogisticRegression(solver='liblinear'))
logistic_model.fit(X_train, Y_train)

# Realizando a classificação no conjunto de validação
previsao_logistic = logistic_model.predict(X_validation)

# Relatório de Classificação
print('Relatório de Classificação - Regressão Logística:\n', classification_report(Y_validation, previsao_logistic))

# Matriz de Confusão
matriz_confusao = pd.crosstab(Y_validation, previsao_logistic, rownames=['Real'], colnames=['Previsto'])
sns.heatmap(matriz_confusao, cmap='coolwarm', annot=True, fmt='d')
plt.title('Matriz de Confusão - Regressão Logística')
plt.show()

# Avaliação da Curva ROC para classificação multiclass
# Binarizando os labels para One-vs-Rest
Y_validation_bin = label_binarize(Y_validation, classes=np.unique(Y_validation))
previsao_logistic_bin = label_binarize(previsao_logistic, classes=np.unique(Y_validation))

fpr = {}
tpr = {}
roc_auc = {}
n_classes = Y_validation_bin.shape[1]

for i in range(n_classes):
    fpr[i], tpr[i], _ = roc_curve(Y_validation_bin[:, i], previsao_logistic_bin[:, i])
    roc_auc[i] = auc(fpr[i], tpr[i])

# Plotando a Curva ROC para cada classe
plt.figure()
for i in range(n_classes):
    plt.plot(fpr[i], tpr[i], label=f'Classe {i+1} (AUC = {roc_auc[i]:.2f})')

plt.plot([0, 1], [0, 1], 'k--')
plt.title('Curva ROC - Regressão Logística (Multiclass)')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend(loc='lower right')
plt.show()
```

```python
# Modelo de Análise Discriminante
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score, roc_curve, auc
from sklearn.preprocessing import label_binarize
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

# Treinando o modelo de Análise Discriminante Linear
lda_model = LinearDiscriminantAnalysis()
lda_model.fit(X_train, Y_train)

# Realizando a classificação no conjunto de validação
previsao_lda = lda_model.predict(X_validation)

# Relatório de Classificação com zero_division para evitar warnings
print('Relatório de Classificação - Análise Discriminante:\n',
      classification_report(Y_validation, previsao_lda, zero_division=0))

# Matriz de Confusão
matriz_confusao = pd.crosstab(Y_validation, previsao_lda, rownames=['Real'], colnames=['Previsto'])
sns.heatmap(matriz_confusao, cmap='coolwarm', annot=True, fmt='d')
plt.title('Matriz de Confusão - Análise Discriminante')
plt.show()

# Avaliação da Curva ROC para classificação multiclass
Y_validation_bin = label_binarize(Y_validation, classes=np.unique(Y_validation))
previsao_lda_bin = label_binarize(previsao_lda, classes=np.unique(Y_validation))

fpr = {}
tpr = {}
roc_auc = {}
n_classes = Y_validation_bin.shape[1]

# Calculando a curva ROC e AUC para cada classe
for i in range(n_classes):
    fpr[i], tpr[i], _ = roc_curve(Y_validation_bin[:, i], previsao_lda_bin[:, i])
    roc_auc[i] = auc(fpr[i], tpr[i])

# Plotando a Curva ROC para cada classe
plt.figure()
for i in range(n_classes):
    plt.plot(fpr[i], tpr[i], label=f'Classe {i+1} (AUC = {roc_auc[i]:.2f})')

plt.plot([0, 1], [0, 1], 'k--')
plt.title('Curva ROC - Análise Discriminante Linear (Multiclass)')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend(loc='lower right')
plt.show()

```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans, DBSCAN, MeanShift
from sklearn.decomposition import PCA
from sklearn.metrics import silhouette_score

scaler = StandardScaler()
X_scale = scaler.fit_transform(X)

# Modelos não supervisionados a serem testados
models = []
models.append(KMeans(n_clusters=3, random_state=30, n_init="auto"))
models.append(DBSCAN(eps=3, min_samples=10))
models.append(PCA(n_components=3))
models.append(MeanShift())

# Nome dos modelos
nome = ['kmeans', 'DBSCAN', 'PCA', 'MS']


results = []
n_grupos = []


n = 0
for model in models:

    results.append(model.fit(X_scale))

    if nome[n] == 'PCA':

        n_grupos.append(3)
        msg = "%s: %i (não aplicável para PCA)" % (nome[n], 3)
    elif nome[n] == 'MS':  # MeanShift
        labels = model.labels_
        centers = model.cluster_centers_
        n_grupos.append(len(centers))
        silhouette_avg = silhouette_score(X_scale, labels) if len(set(labels)) > 1 else None
        msg = "%s: %i (Silhouette: %s)" % (nome[n], len(centers), f'{silhouette_avg:.4f}' if silhouette_avg is not None else 'N/A')
    else:
        labels = model.labels_ if hasattr(model, 'labels_') else model.fit_predict(X_scale)
        n_grupos.append(len(set(labels)) - (1 if -1 in labels else 0))

        if len(set(labels)) > 1:
            silhouette_avg = silhouette_score(X_scale, labels)
            msg = "%s: %i (Silhouette: %f)" % (nome[n], n_grupos[n], silhouette_avg)
        else:
            msg = "%s: %i (Silhouette: N/A)" % (nome[n], n_grupos[n])

    n += 1
    print(msg)

```

```python
#Gráfico Kmeans
scaler = StandardScaler()
X_scale = scaler.fit_transform(X)

# Aplicando o modelo KMeans (resultado de um modelo previamente ajustado)
kmeans_model = KMeans(n_clusters=3, random_state=30, n_init="auto")
kmeans_model.fit(X_scale)

# Armazenando os resultados
results = [kmeans_model]

# Definindo as cores para os clusters
unique_labels = set(results[0].labels_)
colors = [plt.cm.Spectral(each) for each in np.linspace(0, 1, len(unique_labels))]
core_samples_mask = np.zeros_like(results[0].labels_, dtype=bool)

# Gerando o gráfico
for k, col in zip(unique_labels, colors):
    if k == -1:
        # Black used for noise
        col = [0, 0, 0, 1]

    class_member_mask = (results[0].labels_ == k)

    # Plotando os pontos do cluster
    xy = X_scale[class_member_mask & core_samples_mask]
    plt.plot(
        xy[:, 0],
        xy[:, 1],
        "o",
        markerfacecolor=tuple(col),
        markeredgecolor="k",
        markersize=14,
    )

    xy = X_scale[class_member_mask & ~core_samples_mask]
    plt.plot(
        xy[:, 0],
        xy[:, 1],
        "o",
        markerfacecolor=tuple(col),
        markeredgecolor="k",
        markersize=6,
    )

# Título com o número de grupos formados
plt.title('%d Grupos Formados pelo KMeans' % n_grupos[0])
plt.show()
```
